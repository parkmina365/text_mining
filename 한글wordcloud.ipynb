```python
import pandas as pd
from konlpy.tag import Okt
import re

okt = Okt()

with open('/content/drive/MyDrive/cakd3_colab/textmining/dataset/영화 기생충_review.txt','r',
          encoding = 'cp949') as file:
          text = file.read()
          print(text)
```

```python
def cleanText(readData):
  text = re.sub('[0-9]+','',readData)
  text = re.sub('[A-Za-z]+','',text)
  text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ·!』\\‘’|\(\)\[\]\<\>`\'…》]','',text)
  return text
text = cleanText(text)

morph = okt.pos(text)
word_list = []
for word, tag in morph:
  if (tag == 'Noun') | (tag == 'Verb') | (tag == 'Adjective'):
    word_list.append(word)
print(word_list)
```

```python
from collections import Counter
count = Counter(word_list)
n_words = dict(count.most_common())
n_words
```

```python
import numpy as np
data = pd.Series(n_words)
df=pd.DataFrame(data, columns=['cnt'])
df10 = df.sort_values(by='cnt', ascending=False).head(10)
df10.reset_index(inplace=True)
df10
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
sns.barplot(x='index', y='cnt', data=df10)

### 폰트 깨질시
### 마운트초기화 > 폰트설치 > 런타임다시시작 > 구글드라이브 마운트 > 나머지 코드
```

```python
# Noun만 뽑기

def cleanText(readData):
  text = re.sub('[0-9]+', '', readData)
  text = re.sub('[A-Za-z]+', '', text)
  text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ·!』\\‘’|\(\)\[\]\<\>`\'…》]','',text)
  return text

text = cleanText(text)

morph = okt.pos(text)
noun_list = []

for word, tag in morph:
  if tag == 'Noun':
    noun_list.append(word)
print(noun_list)
```

```python
from collections import Counter
count = Counter(noun_list)
words = dict(count.most_common())
words
```
